## Exploratory Data Analysis

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load in your csv.gz's of results for each year extracted.

```python
# Open 2000 saved file and preview again
df_tmdb_2000 = pd.read_csv("Data/final_tmdb_data_2000.csv.gz", low_memory = False)
df_tmdb_2000.head()
```

```python
# Open 2001 saved file and preview again
df_tmdb_2001 = pd.read_csv("Data/final_tmdb_data_2001.csv.gz", low_memory = False)
df_tmdb_2001.head()
```

### Concatenate the data into 1 data frame for the remainder of the analysis.

```python
df_tmdb_combined = pd.concat([df_tmdb_2000, df_tmdb_2001])
df_tmdb_combined.head()
```

```python
# shape of the concatenated data frame
df_tmdb_combined.shape
```

```python
# make a working copy of the data frame
df_movies = df_tmdb_combined.copy()
df_movies.head()
```

### Exploratory Data Analysis:

#### 1. How many movies had at least some valid financial information (values > 0 for budget OR revenue)?
- Please exclude any movies with 0's for budget AND revenue from the remaining visualizations.

```python
# filter movies with only positive budget OR positive revenue
budget_filter = df_movies['budget'] > 0
revenue_filter = df_movies['revenue'] > 0

df_movies = df_movies.loc[budget_filter | revenue_filter, :]
df_movies.head()
```

```python
# shape of the filtered data frame

df_movies.shape
```

1. **There are 728 movies remaining with at least some financial information. (values > 0 for budget OR revenue.)**

```python
# filter only movies with (BOTH) positive budget AND positive revenue

budget_filter = df_movies['budget'] > 0
revenue_filter = df_movies['revenue'] > 0


df_movies = df_movies.loc[budget_filter & revenue_filter, :]
df_movies.head()
```

```python
# shape of filtered data frame
df_movies.shape
```

#### 2. How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
df_movies['certification'].value_counts()
```

2. **There are 144 R movies, 117 PG-13 movies, 30 PG movies, 13 G movies, and also 6 NR movies.**

#### 3. What is the average revenue per certification category?

```python
df_movies.groupby('certification')['revenue'].mean()
```

```python
## Plot a bar plot of the average revenue for different certifications
sns.barplot(data=df_movies, y='revenue', x='certification',)
```

3. **The average revenue per certification category is 13,321,690 dollars for G movies, 2,383,332 dollars for NR movies, 13,049,040 dollars for PG movies, 11,107,400 dollars for PG-13 movies, and 5,230,737 dollars for R movies.**

#### 4. What is the average budget per certification category?

```python
df_movies.groupby('certification')['budget'].mean()
```

```python
## Plot a bar plot of the budget for different certifications
sns.barplot(data=df_movies, y='budget', x='certification',)
```

3. **The average budget per certification category is 4,400,000 dollars for G movies, 1,478,333 dollars for NR movies, 5,218,333 dollars for PG movies, 4,680,682 dollars for PG-13 movies, and 2,695,496 dollars for R movies.**
